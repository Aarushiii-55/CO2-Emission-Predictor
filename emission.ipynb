### Algorithms

```python
data.info()
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error,accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df_dummies = pd.get_dummies(data=data[['Make','Model','Vehicle_Class','Transmission','Fuel_Type']])
df_num_features = data.select_dtypes(include=np.number)
df_comb = pd.concat([df_num_features,df_dummies],axis=1)
```

```python
df_comb
```

LINEAR REGRESSION

```python
x = df_comb.drop('CO2_Emissions',axis=1)
y =df_comb['CO2_Emissions']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
scalar = StandardScaler()
X_train = scalar.fit_transform(x_train)
X_test = scalar.transform(x_test)
```

```python
model_lin = LinearRegression()
model_lin.fit(X_train,y_train)
```

```python
y_pred_lr = model_lin.predict(X_test)
y_pred_lr
```

```python
mean_squared_error(y_test,y_pred_lr)
```

```python
r2_score(y_test,y_pred_lr)
```

```python

# Frame predicted values
frames = [y_pred_lr, y_test]
result_pred = pd.DataFrame(data=frames)
result_pred = result_pred.T
```

## Model Comparison

```python
models = ['model_lin','model_dtr','randomforest','model_svr']
frame = pd.DataFrame(columns={'models':[],'mse':[], 'mae':[] ,'r2score':[]})
for i in range(len(models)):
    if models[i] =='model_lin':
        model_lin= LinearRegression()
        model_lin.fit(X_train,y_train)
        y_pred=model_lin.predict(X_test)
        mse =mean_squared_error(y_test,y_pred)
        mae = mean_absolute_error(y_test,y_pred)
        r2score = r2_score(y_test,y_pred)
        frame.loc[frame.shape[0]] = ['Linear Regression',mse,mae,r2score]
    elif models[i] == 'model_dtr':
        model_dtr = DecisionTreeRegressor()
        model_dtr.fit(X_train,y_train)
        y_pred=model_dtr.predict(X_test)
        mse =np.sqrt(mean_squared_error(y_test,y_pred))
        mae = mean_absolute_error(y_test,y_pred)
        r2score = r2_score(y_test,y_pred)
        frame.loc[frame.shape[0]] = ['Decision Tree',mse,mae,r2score]
    elif models[i]=='randomforest':
        randomforest = RandomForestRegressor()
        randomforest.fit(X_train,y_train)
        y_pred=randomforest.predict(X_test)
        mse =mean_squared_error(y_test,y_pred)
        mae = mean_absolute_error(y_test,y_pred)
        r2score = r2_score(y_test,y_pred)
        frame.loc[frame.shape[0]] = ['Random Forest',mse,mae,r2score]
    else:
        models[i]=='model_svr'      
        model_svr = SVR()
        model_svr.fit(X_train,y_train)
        y_pred=model_svr.predict(X_test)
        mse =mean_squared_error(y_test,y_pred)
        mae = mean_absolute_error(y_test,y_pred)
        r2score = r2_score(y_test,y_pred)
        frame.loc[frame.shape[0]] = ['SVR',mse,mae,r2score]
        
           



        
```
